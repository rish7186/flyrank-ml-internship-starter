## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

Ranked Actions

Priority 1:
Review districts with high median income, a large number of households, and newer housing first. These areas received the highest baseline scores and are the strongest candidates for higher house values.

Priority 2:
Review districts that satisfy two of the three baseline conditions. These districts show moderate potential and should be evaluated after the highest-priority group.

Priority 3:
Review the remaining districts if additional analysis is needed. These districts have lower baseline scores and less evidence that they represent high-value housing areas.

Reason Codes:
- high_income: The district has above-median household income.
- many_households: The district has above-median household count.
- newer_housing: The housing median age is below the dataset median.
- strong_candidate: Multiple positive conditions are satisfied, giving the district a higher priority score.

These recommendations are intended to support decision-making and prioritization rather than provide definitive predictions.

```python
import pandas as pd # Import pandas library
import os # Import os module for directory operations
# Assuming df needs to be loaded from a CSV file.
# Please replace 'path/to/your/data.csv' with the actual path to your dataset.
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Calculate medians for comparison to create 'baseline_score' and 'reason_code'
median_income_median = df['median_income'].median()
households_median = df['households'].median()
housing_median_age_median = df['housing_median_age'].median()

# Create boolean flags for conditions
df['is_high_income'] = df['median_income'] > median_income_median
df['is_many_households'] = df['households'] > households_median
df['is_newer_housing'] = df['housing_median_age'] < housing_median_age_median

# Calculate baseline_score as the sum of true conditions
df['baseline_score'] = df['is_high_income'].astype(int) + \
                      df['is_many_households'].astype(int) + \
                      df['is_newer_housing'].astype(int)

# Create reason_code based on the conditions and baseline_score
def get_reason_code(row):
    reasons = []
    if row['is_high_income']:
        reasons.append('high_income')
    if row['is_many_households']:
        reasons.append('many_households')
    if row['is_newer_housing']:
        reasons.append('newer_housing')

    if row['baseline_score'] == 3:
        return 'strong_candidate: ' + ', '.join(reasons)
    elif reasons:
        return ', '.join(reasons)
    else:
        return 'no_specific_criteria_met'

df['reason_code'] = df.apply(get_reason_code, axis=1)

# Drop intermediate boolean columns as they are no longer needed
df = df.drop(columns=['is_high_income', 'is_many_households', 'is_newer_housing'])

# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Display the ranked action queue

queue = df[[
    "baseline_score",
    "reason_code",
    "median_income",
    "households",
    "housing_median_age"
]].sort_values(
    by="baseline_score",
    ascending=False
)

print("Top 10 Ranked Actions")
display(queue.head(10))

# Create the directory if it doesn't exist
os.makedirs('work/outputs', exist_ok=True)

# Optional: Save the ranked queue
queue.to_csv("work/outputs/ranked_actions.csv", index=False)

print("Ranked action queue saved to work/outputs/ranked_actions.csv")
```
